### Functionalize

```python
def calcLoad(curIn,curH):
    curLoad = 0
    for y,r in enumerate(curIn):
        toAdd = np.sum([x=='O' for x in r])
        curLoad += toAdd * (curH-y)
        
    return curLoad
```

```python
def moveUp(curPos,curW):
    # Move all upwards
    for y,r in enumerate(curPos):
        for x in range(curW):
            # Check if 'O' 
            if r[x] == 'O':
                # Move upwards
                yOld = y 
                y2 = yOld - 1
                while y2 >= 0:
                    if curPos[y2][x] == '.':
                        curPos[y2] = curPos[y2][:x] + 'O' + curPos[y2][x+1:]
                        curPos[yOld] = curPos[yOld][:x] + '.' + curPos[yOld][x+1:]

                        yOld = y2
                        y2 -= 1
                    else:
                        # Stop while-loop
                        y2 = -1
    return curPos 
```

```python
curIn = ogIn.copy()
curH = len(curIn)
curW = len(curIn[0])

curIn = moveUp(curIn,curW)
calcLoad(curIn,curH)
```

```python
def moveDown(curPos,curW,curH):
    # Move all downwards
    # for y,r in enumerate(curPos):
    for y in range(curH-1,-1,-1):
        r = curPos[y]
        for x in range(curW):
            # Check if 'O' 
            if r[x] == 'O':
                # Move downwards
                yOld = y 
                y2 = yOld + 1
                while y2 < curH:
                    if curPos[y2][x] == '.':
                        curPos[y2] = curPos[y2][:x] + 'O' + curPos[y2][x+1:]
                        curPos[yOld] = curPos[yOld][:x] + '.' + curPos[yOld][x+1:]

                        yOld = y2
                        y2 += 1
                    else:
                        # To stop while-loop
                        y2 = curH
    return curPos 
```

```python
def moveLeft(curPos,curW):
    # Move all leftwards
    for y,r in enumerate(curPos):
    # for y in range(curH-1,-1,-1):
    #     r = curPos[y]
        for x in range(curW):
            # Check if 'O' 
            if r[x] == 'O':
                # Move left
                xOld = x
                x2 = xOld - 1
                while x2 >= 0:
                    if curPos[y][x2] == '.':
                        curPos[y] = curPos[y][:x2] + 'O.' + curPos[y][x2+2:]

                        xOld = x2
                        x2 -= 1
                    else:
                        # To stop while-loop
                        x2 = -1
    return curPos 
```

```python
def moveRight(curPos,curW):
    # Move all rightwards
    for y,r in enumerate(curPos):
        # for x in range(curW):
        for x in range(curW-1,-1,-1):
            # Check if 'O' 
            if r[x] == 'O':
                # Move right
                xOld = x
                x2 = xOld + 1
                while x2 < curW:
                    if curPos[y][x2] == '.':
                        curPos[y] = curPos[y][:x2-1] + '.O' + curPos[y][x2+1:]

                        xOld = x2
                        x2 += 1
                    else:
                        # To stop while-loop
                        x2 = curW
    return curPos 
```

```python

cardNames = ['A' , 'K', 'Q', 'J', 'T', '9', '8', '7', '6', '5', '4', '3', '2']
cardDict = dict(zip(cardNames,np.arange(len(cardNames),0,-1)) )
```

```python
htypes = ['five','four','house','three','twopair','onepair','high']
hvals = np.arange(7,0,-1)
typeTxt = dict(zip(hvals,htypes))
```

```python
def determineType(hand):
    # print(hand)
        
    hand = [x for x in hand]

    uni,counts = np.unique(hand,return_counts=True)

    if 5 in counts:
        # return 'five'
        return 7
    elif 4 in counts:
        # return 'four'
        return 6
    elif 3 in counts:
        if 2 in counts:
            # return 'house'
            return 5
        else:
            # return 'three'
            return 4
    elif 2 in counts:
        # Contains at least one pair 
        numPairs = (2==counts).sum()
        if numPairs == 2:
            # return 'twopair'
            return 3
        if numPairs == 1:
            # return 'onepair'
            return 2
    else:
        # return 'high'
        return 1
```

```python
def compareHandsFast(hand1,hand2,ht1,ht2):
    # print('Comparing ',hand1,hand2)
    # ht1 = determineType(hand1)
    # ht2 = determineType(hand2)

    # print(typeTxt[ht1],typeTxt[ht2])
    if ht1 > ht2:
        # print(hand1,'wins')
        # return (hand1,hand2)
        return 1
    elif ht1 < ht2:
        # print(hand2,'wins')
        # return (hand2,hand1)
        return 2
    else:
        # Go through cards one at a time
        for i in range(5):
            c1 = hand1[i]
            c2 = hand2[i]
            # print(c1,c2)
            if cardDict[c1] > cardDict[c2]:
                # print(c1,'wins over',c2)
                # return (hand1,hand2)
                return 1
            elif cardDict[c1] < cardDict[c2]:
                # print(c1,'loses to',c2)
                # return (hand2,hand1)
                return 2
```

```python
curIn = inputs 
# curIn = inputsTest

curIn
hands = []
bids = []
for li in curIn:
    ch,cb = li.split()

    hands.append(ch)
    bids.append(int(cb))
    
typeHands = [determineType(h) for h in hands]
```

```python
# Bubble sort time
# hands = iniHands.copy()
# bids = iniBids.copy()

numPass = 0
isSorted = False
# for k in range(20):
while isSorted == False:
    numSwaps = 0
    for i in range(len(hands)-1):
        h1 = hands[i]
        h2 = hands[i+1]
        b1 = bids[i]
        b2 = bids[i+1]
        ht1 = typeHands[i]
        ht2 = typeHands[i+1]

        # print(h1,h2)
        # curWin = compareHands(h1,h2)
        curWin = compareHandsFast(h1,h2,ht1,ht2)
        # print(curWin)
        if curWin == 1:
            numSwaps += 1
            hands[i] = h2
            hands[i+1] = h1
            bids[i] = b2
            bids[i+1] = b1
            typeHands[i] = ht2
            typeHands[i+1] = ht1

    # Make a final extra comparison of the last two
    if numSwaps == 0:
        isSorted = True
    
    numPass += 1
    if (numPass % 100) == 0:
        print('Pass:',numPass,' and number of swaps in this passs',numSwaps)

# hands
# bids
```

```python
bidsRank = np.arange(len(bids)) + 1
sol1 = np.sum(bids * bidsRank)
sol1 
```
